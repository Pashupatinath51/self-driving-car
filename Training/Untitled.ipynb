```python
import cv2
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


import matplotlib.image as mpimg
```

```python
def preProcess(img):
    img = img[54:120,:,:]
    img = cv2.cvtColor(img, cv2.COLOR_RGB2YUV)
    img = cv2.GaussianBlur(img,  (3, 3), 0)
    img = cv2.resize(img, (200, 66))
    img = img/255
    return img

imgRe = preProcess(mpimg.imread(r'C:\Users\shah9\Desktop\Project\Training\DataCollected\IMG1/Image_1649430038432.jpg'))
mpimg.imsave('Result.jpg',imgRe)
plt.imshow(imgRe)
plt.show()
```
